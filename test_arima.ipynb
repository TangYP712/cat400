```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sales_5_6.csv')
df
```

```python
df.drop(columns='sales_id', inplace=True)
df.head()
```

```python
df2 = df
df2.columns = ['Date','Sales']
df2
```

```python
df3 = df2.set_index('Date')
df3
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def ad_test(dataset):
    dftest = adfuller(dataset, autolag="AIC")
    print("1. ADF: ",dftest[0])
    print("2. P-Value: ",dftest[1])
    print("3. Num of Lags: ",dftest[2])
    print("4. Num of Observations Used for ADF Regression and Critical Values Calculation: ",dftest[3])
    print("5. Critical Values: ")
    for key, val in dftest[4].items():
        print("\t",key,": ",val)
```

```python
ad_test(df3['Sales'])
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
print(df3.shape)
train = df3.iloc[:-20]
test = df3.iloc[-20:]
print(train.shape, test.shape)
```

```python
model = ARIMA(train['Sales'], order=(0,0,1))
model = model.fit()
model.summary()
```

```python
start = len(train)
end = len(train)+len(test)-1
pred = model.predict(start=start,end=end,typ='levels')
print(pred)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse = sqrt(mean_squared_error(pred, test['Sales']))
print(rmse)
```

```python
rmse2 = np.mean((pred.values-test['Sales'])**2)**.5
print(rmse2)
```

```python
corr = np.corrcoef(pred.values-test['Sales'])   # Correlation between the Actual and the Forecast (corr)
mae = np.mean(np.abs(pred.values-test['Sales']))    # Mean Absolute Error
mpe = np.mean((pred.values-test['Sales'])/test['Sales'])   # Mean Percentage Error
print(corr)
print(mae)
print(mpe)
```

```python
test['Sales'].mean()
```

```python
mape = np.mean(np.abs(pred.values-test['Sales'])/np.abs(test['Sales']))  # Mean Absolute Percentage Error (MAPE)
me = np.mean(pred.values-test['Sales'])             # Mean Error
print(mape)
print(me)
```

```python
normalized_rmse = rmse/(test['Sales'].max()-test['Sales'].min())
print(normalized_rmse)
```
